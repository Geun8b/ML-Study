**[독립변수 / 특성 값 / 문제 데이터]**

'CRIM' : 구역별 1인당 범죄율 - 범죄율이 높으면 집 값이 낮아지는 경향

'ZN' : 25,00 평방피트를 초과하는 거주지역의 비율 - 대여 주택 단지 비중

'INDUS' : 구역별 비소매 상업지역 점유 비율 - 공업, 상업 시설 밀집도

'CHAS' : 찰스강과 인접해있는지 (0 = 인접X / 1 = 인접O) - 조망권 / 입지 가치

'NOX' : 대기 중 일산화질소 농도 - 환경 오염 정도

'RM' : 주택 1가구당 평균 방의 갯수

'AGE' : 1940년 이전 건축 된 소유주택의 비율

'DIS' : 보스턴 직업센터까지의 가중 거리 - 도심 접근성

'RAD' : 방사형 고속도로 접근성 지수 - 교통 편의성

'TAX' : 10,000달러당 재산세의 비율 - 세금 부담 정도

'PTRATIO' : 구역별 학생, 교사 비율 - 교육 환경(낮을 수록 학급당 학생 수가 적음)

'B' : 흑인의 비율

'LSTAT' : 하위 계층의 비율 - 소득, 교육, 직종 등

**[족속변수 / 예측 값 / 정답 데이터]**

'MEDV' : 본인 소유 주택 가격의 중앙 값(단위는 $1000)

'CAT.MEDV' : 30,000달러 이상이면 1 / 미만이면 0
